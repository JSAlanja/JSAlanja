```python
import cv2
import numpy as np

# Load the pre-trained Haar Cascade Classifier for face detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load the trained LBPH face recognizer (make sure you've trained it previously)
recognizer = cv2.face_LBPHFaceRecognizer_create()
recognizer.read("D:/Alan ja/darwin.j.jpg")
# Authorized users' labels (e.g., '1' for John, '2' for Sarah)
authorized_users = {1: "John", 2: "Sarah"}

# Start video capture (e.g., webcam)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for (x, y, w, h) in faces:
        face = gray[y:y + h, x:x + w]
        
        # Recognize the face
        label, confidence = recognizer.predict(face)
        
        # Check if recognized face is in the authorized users list
        if label in authorized_users:
            name = authorized_users[label]
            cv2.putText(frame, f'Welcome {name}', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            # Trigger access (e.g., unlock door, grant access)
            print(f"Access Granted to {name}")
            # You can add code here to unlock a door, etc.
        else:
            cv2.putText(frame, 'Unknown Face', (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
            # Trigger alarm or log the event
            print("Intruder detected! Alert!")
            # Trigger security action (e.g., send alert or turn on alarm)
        
        # Draw a rectangle around the detected face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
    
    # Show the frame
    cv2.imshow('Security System', frame)
    
    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close windows
cap.release()
cv2.destroyAllWindows()
```
